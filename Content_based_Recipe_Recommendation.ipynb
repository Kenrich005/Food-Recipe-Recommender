```python
df = pd.read_csv("https://media.githubusercontent.com/media/pankajsr/data-sets/master/AMPBA/MLUL2/recipes_with_nutritional_info.csv",
                usecols = ["title","nutr_values_per100g","ingredients","instructions","url"])
df.shape
```

```python
df.info()
```

```python
df.head()
```

```python
# Check if any Title occurs more than once
df.title.value_counts()
```

```python
df['nutr_values_per100g'] = df['nutr_values_per100g'].apply(lambda row: ast.literal_eval(row))

df2 = pd.json_normalize(df['nutr_values_per100g'])
df2.head()
```

```python
df2.iloc[10]
```

```python
df3 = df['ingredients'].apply(lambda row: ast.literal_eval(row))

df3 = df3.apply(lambda row: [dict1['text'] for dict1 in row])

df3 = df3.apply(lambda row: ' '.join(map(str, row)))

df3.head()
```

```python
df4 = df['instructions'].apply(lambda row: ast.literal_eval(row))

df4 = df4.apply(lambda row: [dict1['text'] for dict1 in row])

df4 = df4.apply(lambda row: ' '.join(map(str, row)))

df4.head()
```

```python
# Creating cleaned data
```

```python
df.head()
```

```python
df_cleaned = pd.concat([df.title, df2, df3, df4, df.url],axis=1)
df_cleaned.head()
```

```python
df_cleaned.info()
```

## KNN Modelling

```python
from sklearn.neighbors import NearestNeighbors
model_knn = NearestNeighbors(metric='cosine', algorithm='brute', n_neighbors=20, n_jobs=-1)

model_knn.fit(df2)
```

```python
distances, indices = model_knn.kneighbors(pd.DataFrame(df2.iloc[1000]).T, n_neighbors=7)
```
